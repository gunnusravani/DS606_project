```python
import os
import pandas as pd
from tqdm import tqdm
import time
from googletrans import Translator
from nltk.tokenize import sent_tokenize

# def extract_language_from_path(path):
#     parts = path.split(os.sep)
#     return parts[-2] if len(parts) >= 2 else None
def extract_language_from_path(file_path):
    """
    Extract language name from a file path like:
    openai_jailbreak_bengali.csv  -> 'bengali'
    """
    filename = os.path.basename(file_path)
    name_without_ext = os.path.splitext(filename)[0]
    
    parts = name_without_ext.lower().split("_")
    
    # Known languages (extend as needed)
    known_languages = {
        "english", "hindi", "bengali", "telugu",
        "marathi", "tamil", "urdu", "gujarati",
        "kannada", "malayalam", "punjabi"
    }
    
    for part in parts:
        if part in known_languages:
            return part
    
    return None
def map_language_name_to_code(lang_name):
    mapping = {
        "hindi": "hi",
        "bengali": "bn",
        "telugu": "te",
        "marathi": "mr",
        "english": "en"
    }
    return mapping.get(lang_name.lower(), None)



def translate_sentence_safe(sentence, src_lang, translator, max_retries=2):
    sentence = str(sentence)

    def try_translate(s, src):
        for attempt in range(max_retries):
            try:
                translated = translator.translate(s, src=src, dest="en")
                return translated.text if translated else None
            except Exception as e:
                print(f"Error translating with src={src} (attempt {attempt+1}): {e}")
                time.sleep(1.5 * (attempt + 1))
        return None

    # First try full sentence
    result = try_translate(sentence, src_lang)

    # Fallback: try with src='en'
    if result is None and src_lang != "en":
        result = try_translate(sentence, "en")

    # Still failing? Try splitting into two sentences
    if result is None:
        print("🔁 Attempting sentence split and translation...")
        try:
            sentences = sent_tokenize(sentence)
            if len(sentences) >= 2:
                first_half = " ".join(sentences[:len(sentences)//2])
                second_half = " ".join(sentences[len(sentences)//2:])
                
                first_trans = try_translate(first_half, src_lang) or try_translate(first_half, "en")
                second_trans = try_translate(second_half, src_lang) or try_translate(second_half, "en")

                if first_trans and second_trans:
                    result = first_trans + " " + second_trans
                elif first_trans:
                    result = first_trans + " " + second_half
                elif second_trans:
                    result = first_half + " " + second_trans


                else:
                    print(f"❌ Split translation failed:\n{sentence[:150]}...")
                    result = None
                print("result sent successfully")
#                 print(result)
            else:
                print("⚠️ Could not split into two sentences. Skipping.")
        except Exception as e:
            print(f"❌ Sentence splitting error: {e}")
            result = None

    return result

```

```python
def translate_columns(file_path, translator, column_pairs):
    try:
        df = pd.read_csv(file_path)
        col = df.columns

        # Rename fallback: if some expected cols aren't found but alternatives exist
        if "trans_response" in col and "response" not in col:
            df = df.rename(columns={"trans_response": "response"})
            print("Column renamed: trans_response → response")

        lang_folder = extract_language_from_path(file_path)
        src_lang = map_language_name_to_code(lang_folder)

        if not src_lang:
            print(f"⚠️ Skipping {file_path}: Unknown language '{lang_folder}'")
            return

        print(f"🔵 Processing file: {file_path} | Source language: {src_lang}")

        for src_col, tgt_col in column_pairs:
            if src_col not in df.columns:
                print(f"⚠️ Skipping column {src_col} in {file_path}: Column not found.")
                continue

            if tgt_col not in df.columns:
                df[tgt_col] = ""

            needs_translation = (
                df[tgt_col].isnull() | (df[tgt_col].astype(str).str.strip() == "")
            ) & (df[src_col].astype(str).str.strip() != "")

            rows_to_translate = df[needs_translation].copy()

            if not rows_to_translate.empty:
                print(f"🔄 Translating {len(rows_to_translate)} rows for column '{src_col}'...")
                translated_texts = []

                for response in tqdm(rows_to_translate[src_col], desc=f"Translating {src_col}", ncols=100):
                    translated_text = translate_sentence_safe(response, src_lang, translator)
                    translated_texts.append(translated_text if translated_text is not None else response)

                df.loc[needs_translation, tgt_col] = translated_texts
            else:
                print(f"✅ No rows need translation in column '{src_col}'.")

        df.to_csv(file_path, index=False)
        print(f"✅ File saved: {file_path}")

    except Exception as e:
        print(f"❌ Error processing {file_path}: {e}")

def process_directory(root_dir):
    translator = Translator()
    column_pairs = [
         ("response_no_context", "gtrans_response_no_context"),
        ("response_with_context", "gtrans_response_with_context")
    
    ]

    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith(".csv") and "checkpoint" not in filename.lower():
                file_path = os.path.join(dirpath, filename)
                translate_columns(file_path, translator, column_pairs)
            else:
                if "checkpoint" in filename.lower():
                    print(f"⏩ Skipping checkpoint file: {filename}")

if __name__ == "__main__":
    root_directory = "./translated_jailbreak_tmlr/attention"
    process_directory(root_directory)
```

```python
# ============================================================
# Robust Re-Translation Pipeline for Failed GoogleTrans Rows
# ============================================================

import os
import re
import time
import pandas as pd
from tqdm import tqdm
from googletrans import Translator

# ------------------------------------------------------------
# 1) Helpers: missing values & script detection
# ------------------------------------------------------------

INDIC_SCRIPT_RE = re.compile(
    r"[\u0900-\u097F\u0980-\u09FF\u0A00-\u0A7F\u0A80-\u0AFF"
    r"\u0B00-\u0B7F\u0B80-\u0BFF\u0C00-\u0C7F\u0C80-\u0CFF\u0D00-\u0D7F]"
)

def is_missing(x) -> bool:
    if x is None:
        return True
    if isinstance(x, float) and pd.isna(x):
        return True
    if isinstance(x, str) and not x.strip():
        return True
    return False

def contains_indic_script(s: str) -> bool:
    if not isinstance(s, str):
        return False
    return bool(INDIC_SCRIPT_RE.search(s))

def needs_retranslate(src_text, tgt_text) -> bool:
    """
    Decide if a translation is likely bad / missing.
    """
    if is_missing(src_text):
        return False
    if is_missing(tgt_text):
        return True

    src = str(src_text).strip()
    tgt = str(tgt_text).strip()

    if not tgt:
        return True
    if tgt == src:
        return True
    if contains_indic_script(tgt):
        return True

    return False

# ------------------------------------------------------------
# 2) Translation with retries (src='auto')
# ------------------------------------------------------------

def translate_retry(
    text,
    translator,
    dest="en",
    max_retries=4,
    sleep_base=1.0
):
    """
    Retry-safe translation using auto language detection.
    Returns translated text or original on repeated failure.
    """
    if not isinstance(text, str):
        text = "" if pd.isna(text) else str(text)
    
    text = text.strip()
    if not text:
        return text
    
    for attempt in range(max_retries):
        try:
            out = translator.translate(
                text,
                src="auto",
                dest=dest
            ).text

            if out and out.strip():
                return out

            raise RuntimeError("Empty translation")

        except Exception:
            time.sleep(sleep_base * (attempt + 1))

    # return text  # fallback: keep original if all retries fail

# ------------------------------------------------------------
# 3) Re-translate bad rows in a single CSV
# ------------------------------------------------------------

def retranslate_bad_rows_in_file(
    file_path: str,
    column_pairs,
    save_every: int = 200,
    sleep_every: int = 50,
    sleep_seconds: float = 1.0
):
    print(f"\n📄 Processing: {file_path}")
    df = pd.read_csv(file_path)

    # Backward compatibility rename
    if "trans_response" in df.columns and "response" not in df.columns:
        df = df.rename(columns={"trans_response": "response"})

    if "gtrans_retry_flag" not in df.columns:
        df["gtrans_retry_flag"] = ""

    translator = Translator()
    total_retried = 0

    for src_col, tgt_col in column_pairs:
        if src_col not in df.columns:
            print(f"  ⚠️ Missing source column: {src_col}")
            continue

        if tgt_col not in df.columns:
            df[tgt_col] = ""

        mask = df.apply(
            lambda r: needs_retranslate(r[src_col], r[tgt_col]),
            axis=1
        )
        idxs = df.index[mask].tolist()

        if not idxs:
            print(f"  ✅ No retries needed for {src_col} → {tgt_col}")
            continue

        print(f"  🔁 Retranslating {len(idxs)} rows for {src_col} → {tgt_col}")

        for k, i in enumerate(
            tqdm(idxs, desc=f"Retranslating {tgt_col}", leave=False)
        ):  
            print(i)
            src_text = df.at[i, src_col]
            # print(f"index:{i} ")
            new_tgt = translate_retry(src_text, translator)
 
            df.at[i, tgt_col] = new_tgt
            df.at[i, "gtrans_retry_flag"] = "retried"
            total_retried += 1

            if (k + 1) % sleep_every == 0:
                time.sleep(sleep_seconds)

            if save_every and (k + 1) % save_every == 0:
                df.to_csv(file_path, index=False)

        df.to_csv(file_path, index=False)

    print(f"  ✅ Done. Total retried rows: {total_retried}")

# ------------------------------------------------------------
# 4) Run on entire directory (recursive)
# ------------------------------------------------------------

def retranslate_directory(root_dir: str, column_pairs):
    print(f"\n🚀 Starting re-translation in: {root_dir}")
    for dirpath, _, filenames in os.walk(root_dir):
        for fn in filenames:
            if not fn.endswith(".csv"):
                continue
            if "checkpoint" in fn.lower():
                continue

            fp = os.path.join(dirpath, fn)
            retranslate_bad_rows_in_file(fp, column_pairs)

# ------------------------------------------------------------
# 5) CONFIG + RUN
# ------------------------------------------------------------

COLUMN_PAIRS = [
    ("response_no_context", "gtrans_response_no_context"),
    ("response_with_context", "gtrans_response_with_context"),
]

ROOT_DIR = "./translated_jailbreak_tmlr_final/attention"

retranslate_directory(ROOT_DIR, COLUMN_PAIRS)

# ============================================================
# End of pipeline
# ============================================================
```
